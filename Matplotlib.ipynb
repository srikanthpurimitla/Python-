```python
fig,ax=plt.subplots(2,2,figsize=(20,15),linewidth=10,edgecolor='b')
species=['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
box_data=[]
box_plot1=[]
box_plot2=[]
box_plot3=[]
for specie in species:
    box_data.append(df.loc[df['species']==specie]['sepal_length'])
    box_plot1.append(df.loc[df['species']==specie]['sepal_width'])
    box_plot2.append(df.loc[df['species']==specie]['petal_length'])
    box_plot3.append(df.loc[df['species']==specie]['petal_width'])
ax[0][0].boxplot(box_data)
ax[0][0].set_ylabel('sepal_length')
ax[0][0].set_facecolor('#e1ddbf')
ax[0][1].boxplot(box_plot1)
ax[0][1].set_ylabel('sepal_width')
ax[0][1].set_facecolor('#e1ccbf')
ax[1][0].boxplot(box_plot2)
ax[1][0].set_ylabel('petal_length')
ax[1][0].set_facecolor('#e1ccbf')
ax[1][1].boxplot(box_plot3)
ax[1][1].set_ylabel('petal_width')
ax[1][1].set_facecolor('#e1ddbf')
plt.suptitle('Subplots for Iris')
plt.show()
```
